Also im one hot encoding y, so instead of the label just being the number 1 or 2 its an array of 10 elements and the given index has 1 instead of 0. doing this since in the end the result will be a probability for 10 different digits so this way we just align it and get the probability.

```python
X_train = X_train.reshape(60000,28*28)
X_test = X_test.reshape(10000,28*28)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

X_train = X_train/255
X_test = X_test/255
```
